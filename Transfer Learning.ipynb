```python
import numpy as np
import keras
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.layers import Dense, Conv2D, BatchNormalization, Activation
from keras.layers import AveragePooling2D, Input, Flatten
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras.callbacks import ReduceLROnPlateau
from keras.preprocessing.image import ImageDataGenerator
from keras.regularizers import l2
from keras import backend as K
from keras.models import Model
import numpy as np
import os
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.preprocessing.image import ImageDataGenerator
import os
import zipfile
import tensorflow as tf
import numpy as np
import keras
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.layers import Dense, Conv2D, BatchNormalization, Activation
from keras.layers import AveragePooling2D, Input, Flatten
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras.callbacks import ReduceLROnPlateau
from keras.preprocessing.image import ImageDataGenerator
from keras.regularizers import l2
from keras import backend as K
from keras.models import Model
import numpy as np
import os
import tensorflow as tf
# from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l2
from keras.models import Sequential
from keras.layers import Conv2D, Lambda, MaxPooling2D,AveragePooling2D# convolution layers
from keras.layers import Dense, Dropout, Flatten,Activation,Reshape # core layers1
from keras.applications.vgg19 import VGG19
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
from keras import regularizers
from keras.layers import LSTM
from keras.layers import Attention
from keras.layers import Multiply
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Model
from keras.layers import Dense, Conv2D, BatchNormalization, Activation
from keras.layers import AveragePooling2D, Input, Flatten, Reshape, GlobalAveragePooling2D, Dropout
from keras.layers import multiply
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, ReduceLROnPlateau
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.vgg19 import VGG19
from keras import regularizers
from keras.layers import Conv2D, BatchNormalization, Activation, Input, Concatenate
from keras.layers import DepthwiseConv2D, BatchNormalization, Activation, GlobalAveragePooling2D
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, BatchNormalization, Activation, Input, Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.applications import VGG19
from keras.applications import VGG19
from keras.layers import Input, Conv2D, Conv2DTranspose, Concatenate


def grouped_convolution_block(x, filters, groups, kernel_size, strides):
    groups_list = []
    input_groups = tf.split(x, groups, axis=-1)

    for i in range(groups):
        group = Conv2D(filters // groups, kernel_size, strides=strides, padding="same")(input_groups[i])
        groups_list.append(group)

    x = Concatenate(axis=-1)(groups_list)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    return x


def createModel():

    input_shape = (224, 224, 3)
    base_model = VGG19(include_top=False, weights="imagenet", input_shape=input_shape)

    # Freeze the weights of the VGG19 layers
    for layer in base_model.layers:
        layer.trainable = False

    x = base_model.output

    # Apply grouped convolution to the last convolutional layer of VGG19
    x = grouped_convolution_block(x, filters=512, groups=4, kernel_size=3, strides=1)
    x = MaxPooling2D(pool_size=3, strides=2, padding='same')(x)

    # Additional Conv2D layers
    x = Conv2D(512, (3, 3), padding='same')(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=(3, 3), strides=2, padding='same')(x)

    x = Conv2D(256, 3, padding="same")(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=2, strides=2, padding='same')(x)

    x = Conv2D(128, 3, padding="same")(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=2, strides=2, padding='same')(x)

    x = Conv2D(64, 3, padding="same")(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=2, strides=2, padding='same')(x)

    x = Flatten()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.2)(x)
    outputs = Dense(6, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=outputs)
    return model
model=createModel()
```

```python
from tensorflow.keras.optimizers import Adam

# Define the new learning rate
new_lr = 0.001

# Create an instance of the Adam optimizer with the new learning rate
optimizer = Adam(learning_rate=new_lr)

# Compile the model with the updated optimizer
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, min_lr=0.00001, verbose=1)

# Adding callbacks for learning rate reduction and early stopping
#reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)
#early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the combined model
history = model.fit(train_generator, epochs=50, validation_data=val_generator, callbacks=[reduce_lr])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from tensorflow.keras.preprocessing import image

IMAGE_SIZE = (224, 224)  # Adjust according to your model

# Function to predict disease name and severity for a given image
def predict_disease(image_path, model):
    img = image.load_img(image_path, target_size=IMAGE_SIZE)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0  # Normalize the image
    prediction = model.predict(img_array)[0]

    # Interpret predictions based on class mapping
    class_mapping = {
        0: 'bacterial pneumonia',
        1: 'cardiomegaly',
        2: 'lung opacity',
        3: 'normal',
        4: ' tubercolosis',
        5: 'covid-19'
        # Add more mappings as needed based on your classes
    }

    # Get the predicted class
    predicted_class = np.argmax(prediction)
    disease_name = class_mapping[predicted_class]
    confidence = prediction[predicted_class]

    return disease_name, confidence

# Example: Predict disease name and confidence for a random image
random_image_path = '/content/drive/MyDrive/kaggle testin/test/Bacterial pneumonia/124.jpeg'
disease_name, confidence = predict_disease(random_image_path, model)

# Display the image and predictions
img = image.load_img(random_image_path, target_size=IMAGE_SIZE)
plt.imshow(img)
plt.title(f"Predicted Disease: {disease_name}, Confidence: {confidence:.2f}")
plt.show()

```
